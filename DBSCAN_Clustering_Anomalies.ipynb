```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons,make_circles
import matplotlib.pyplot as plt
```

```python
# Generate data
X_circles, _ = make_circles(
    n_samples=750,
    factor=0.3,   # inner circle size
    noise=0.1    # noise level
)

# Plot
plt.figure(figsize=(6,6))
plt.scatter(X_circles[:,0], X_circles[:,1])
plt.title("make_circles Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

### DBSCAN on make_circles

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_circles)
```

```python
# applly DBSCAN
for eps in [0.15,0.17,0.20,0.22,0.24,0.26]:
    db = DBSCAN(eps=eps,min_samples=5)
    labels = db.fit_predict(X_scaled)
    print(eps, set(labels))
```

```python
db = DBSCAN(eps=0.26, min_samples=5)
labels = db.fit_predict(X_scaled)

plt.scatter(X_circles[:,0], X_circles[:,1], c=labels, cmap='viridis')
plt.title("Final DBSCAN Result (Correct)")
plt.show()
```
